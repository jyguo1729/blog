```python
A=[]
with open("latex_custom_command.txt") as file: # Use file to refer to the file object
    cmd=file.readline()
    #print(cmd)
    while(cmd):
        A.append(cmd)
        cmd=file.readline()
        
A
        
```

```python
#      pd: "{\\partial}",
#\newcommand{\sech}{\mathop{\rm sech}\nolimits}
#'\\DeclareMathOperator{\\Rank}{Rank}\n',
import re

def op_replacer(m):
    return '\\newcommand{\\'+m.group(1)+'}{\\mathop{\\\\rm '+m.group(2)+'}\\\\nolimits}'


def cmd_replacer(m):
    return m.group(1)+':\"{\\\\'+m.group(2)+'}\",\n'

C=[]
for x in A:
    C.append(re.sub(r'\\DeclareMathOperator\{\\(\w+)\}\{(\w+)\}\n?',op_replacer,x))
    
C    
B=[]
for x in C:
    B.append(re.sub(r'\\newcommand\{\\(\w+)\}\{\\(.+)\}\n?',cmd_replacer,x))
    
B
```

```python
text="".join(B)
with open("math_jax_macro.txt", "w") as file:
    file.write(text)
```
